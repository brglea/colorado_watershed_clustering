HLS L30 (Harmonized Landsat and Sentinel-2) data, is the combined measurement 
from the NASA/USGS Landsat satellites (8 and 9) and the European Sentinels 
(2A and 2B) (United States Geological Survey (USGS) and NASA 2022). The combination 
of the two is what makes them 'harmonized' and this also means there's more frequent 
observations (images) of the land being taken (every 2-3 days) rather than less 
frequent if used separately (Ju et al. n.d.). The user guide can be found 
[here](https://lpdaac.usgs.gov/documents/1698/HLS_User_Guide_V2.pdf) (Ju et al. n.d.).
The L30 part is in reference to the images being at a special 
30-meter spatial resolution (United States Geological Survey (USGS) and NASA 2022).
These satellites have multiple sensors, each sensor is for a different reflective 
spectral band in the electromagnetic wavelength (Science Education through Earth 
Observation for High Schools (SEOS) and European Association of Remote Sensing 
Laboratories n.d.). A band alone cannot convey much, but the relationship between 
two or more bands does - these are normalized spectral indices such as NDMI, NDVI, 
and others (United States Geological Survey (USGS) n.d.). Information on more in 
detail about remote sensing can be found 
[here](https://seos-project.eu/remotesensing/remotesensing-c01-p06.html) (Science 
Education through Earth Observation for High Schools (SEOS) and European Association 
of Remote Sensing Laboratories n.d.). Bands layered can also output a True Color 
Image that uses the visible bands (red, green, and blue) which is what we typically 
associate with a photo we take with a camera or our phone, and a CIR (Color Infrared) 
Image (a.k.a. False Color Image) which uses different bands (Near Infrared - 'NIR', 
red, and green) (). The use of NIR here enhances the visualization of healthy vegetation 
which the NIR strongly reflects, making healthy vegetation appear bright red - basically 
enhances what the human eye cannot see (). 

To access the data [NASA Worldview](https://worldview.earthdata.nasa.gov) was used 
to search for Denver specifically looking at the HLS L30 dataset (NASA 2025).
A free account will need to be created to access this data and be able to 
use it in your codespace (NASA 2025). The module earthaccess and function search_data 
using parameters are used to search for the HLS L30 data according to 
the parameters set (earthaccess n.d.). Because this data comes from 'professional' 
sources (NASA, USGS, and European agencies) and gathered through satellites, 
there should be a degree of trust in the data, but it depends how you are 
using it and what you are using it for. I am interested in the NDVI to calculate 
or pull vegetation related variables to be able to see if those variables can 
predict % depression prevalence in Denver. 

The data can vary depending on the day or days used (NASA 2025). For example, 
certain days have more or complete cloud coverage and it is essentially useless 
to for this project to pick that single day. It is best to get a 'clear' image 
or images where there is little to no cloud coverage for the best results. I had 
to search for potential clear days that looked to be promising date to use as the 
'temporal' parameter when trying to access this data (will be seen later in the code). 
There will also be differences in the image based on time of year depending 
on what you are looking at. If interested in vegetation, it's 
important to know the climate and environment because you would want a time 
of year where the vegetation is healthiest and would have the greatest % 
reflection; however, it depends on what you are using this data to accomplish. 
In this case study a general vibrant time of year for vegetation like summer 
time or fall in Denver is ideal because the reflectance for vegetation in urban 
greenspaces would be high at this time in theory. Therefore, it would be more 
clear where these green spaces are to get a clear image. Singular dates can be chosen 
or a series of dates or time range for the temporal (datetime) parameter. Choosing a series of 
dates and esentially layering them to get 100% coverage is what will be done here.

The data provided on NASA Worldview can be used for many things beyond what this 
case study aims to do, and it is worth exploring (NASA 2025).
